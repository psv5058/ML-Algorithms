# EXP 5 : To Implement Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd

df=pd.read_csv("C:\\Users\honey\Desktop\Class time\ML\Startups.csv")

print(df)
```

```python
#CLASSIFYING DEPENDENT AND INDEPENDENT VARIABLES

onehotencoder_df = pd.get_dummies(df,columns=['State'])

print(onehotencoder_df)
```

```python
#REMOVE MULTICOLLINEARITY

onehotencoder_df = onehotencoder_df.drop(columns = ['State_California'],axis = 1)

print(onehotencoder_df)
```

```python
#CONSIDERING DEPENDENT AND INDEPENDENT VARIABLES

features = ['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']
x = onehotencoder_df.loc[:,features]
y = onehotencoder_df.loc[:,'Profit']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 10)
```

```python
#FITTING DATA

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
#ACCURACY

print("Train score",regressor.score(x_train,y_train))
print("Train score",regressor.score(x_train,y_train))
```

```python
#MSE & R2 SCORE

from sklearn.metrics import mean_squared_error,r2_score

print(np.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```
