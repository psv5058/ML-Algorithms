## EXP 2: Write a Program to Demonstrate Various Visualization Techniques

```python
## LINE CHART

import matplotlib.pyplot as plt 
x=[10,20,30,40,50,60] 
y=[3,6,5,2,9,11] 
fig = plt.figure(figsize =(8,6))
plt.xlabel("X-axis") 
plt.ylabel("Y-axis") 
plt.title("LINE CHART") 
plt.plot(x,y)
plt.show()
```

```python
## BAR PLOT

import matplotlib.pyplot as plt 
fig = plt.figure(figsize =(7,5))
x=['A','B','C','D','E'] 
y=[97,58,45,68,23] 
plt.xlabel("Students") 
plt.ylabel("Marks") 
plt.title("BAR CHART") 
plt.bar(x,y,color="blue",width=0.3)
plt.show()
```

```python
## SCATTER PLOT 

import matplotlib.pyplot as plt 
fig = plt.figure(figsize =(5,4))
x=[12,45,6,34,98,76,54,34,56,43,21,57] 
y=[23,54,65,7,6,12,3,43,33,12,34,22] 
fig = plt.figure(figsize =(8,5.79))
plt.xlabel("X-axis") 
plt.ylabel("Y-axis") 
plt.title("SCATTER PLOT") 
plt.scatter(x,y,color="magenta") 
plt.show()
```

```python
## HISTOGRAM 

import matplotlib.pyplot as plt 
import numpy as np 
x = np.random.normal(170, 10, 250) 
fig = plt.figure(figsize =(8,5))
plt.xlabel("X-axis") 
plt.ylabel("Y-axis") 
plt.title("HISTOGRAM") 
plt.hist(x) 
plt.show()
```

```python
## PIE CHART

import matplotlib.pyplot as plt 
import numpy as np 
x =np.array([35, 25, 25, 15]) 
mylabels = ["A", "B", "C", "D"] 
mycolors = ["blue", "hotpink", "green", "red"] 
explode=[0,0.1,0,0]
plt.title("PIE CHART")
plt.pie(x, explode=explode,labels = mylabels,colors=mycolors) 
plt.legend(title="Students:",bbox_to_anchor=(1.25,1.25))
plt.show() 
```

```python
## BOX PLOT

import matplotlib.pyplot as plt 
import numpy as np 
np.random.seed(10) 
data = np.random.normal(100, 20, 200) 
fig = plt.figure(figsize =(5,6))
plt.title("BOX PLOT")
plt.xlabel("X-axis") 
plt.ylabel("Y-axis")
plt.boxplot(data) 
plt.show()
```
