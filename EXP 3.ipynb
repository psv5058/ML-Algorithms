## EXP 3: To Plot The Graphs for Employee Database Using Matplotlib

```python
## SCATTER PLOT

import pandas as pd 
import matplotlib.pyplot as plt 
df=pd.read_csv("employee.csv") 
fig = plt.figure(figsize =(10,5))
plt.scatter(df["MANAGER_ID"],df["SALARY"],color="purple") 
plt.title("Scatter plot between Manager_id & Salary") 
plt.xlabel("Manager_id") 
plt.ylabel("Salary") 
plt.show() 
```

```python
## BAR CHART

import pandas as pd 
import matplotlib.pyplot as plt 
df=pd.read_csv("employee.csv") 
fig = plt.figure(figsize =(12,8))
plt.bar(df["DEPARTMENT_ID"],df["JOB_ID"],color="magenta",linewidth="10") 
plt.title("Bar chart between Department_id & Job_id") 
plt.xlabel("Department_id") 
plt.ylabel("Job_id") 
plt.show() 
```

```python
## LINE CHART

import pandas as pd 
import matplotlib.pyplot as plt 
df=pd.read_csv("employee.csv") 
fig = plt.figure(figsize =(15,7))
plt.plot(df["MANAGER_ID"],df["SALARY"],color="green") 
plt.title("Line between Manager_id & Salary") 
plt.xlabel("Manager_id") 
plt.ylabel("Salary") 
plt.show()
```

```python
## HISTOGRAM

import pandas as pd 
import matplotlib.pyplot as plt 
df=pd.read_csv("employee.csv")
fig = plt.figure(figsize =(12,6))
plt.title("Histogram between Employs & Salary") 
plt.xlabel("Salary") 
plt.ylabel("Employee")
plt.hist(df["SALARY"],color="blue")
```

```python
## PIE CHART

import pandas as pd 
import matplotlib.pyplot as plt 
df=pd.read_csv("employee.csv") 
x=df["SALARY"].head(5) 
fig = plt.figure(figsize =(9,5))
plt.title("Pie chart of Salaries")
labels=['A','B','C','D','E'] 
plt.pie(x,labels=labels) 
plt.legend(title="Employee:", bbox_to_anchor=(1.10,1.0)) 
plt.show()
```
